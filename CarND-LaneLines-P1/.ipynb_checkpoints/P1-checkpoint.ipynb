## Import packages

````python
import math
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2
import helper
%matplotlib inline
````

## Read all the Images

````python
image1 = mpimg.imread('test_images/solidWhiteCurve.jpg')
image2 = mpimg.imread('test_images/solidWhiteRight.jpg')
image3 = mpimg.imread('test_images/solidYellowCurve.jpg')
image4 = mpimg.imread('test_images/solidYellowCurve2.jpg')
image5 = mpimg.imread('test_images/solidYellowLeft.jpg')
image6 = mpimg.imread('test_images/whiteCarLaneSwitch.jpg')
image = [image1,image2,image3,image4,image5,image6]

plt.figure()
plt.title('Original')
plt.imshow(image1)
plt.savefig('process/original.png')
````

## Color Selection

````python
color_Select = []

red_threshold = 220
green_threshold = 180
blue_threshold = 0
    
for i in image:
    temp = helper.colorselect(i, red_threshold, green_threshold, blue_threshold)
    color_Select.append(temp)

plt.figure()
plt.title('Color Select')
plt.imshow(color_Select[0])
plt.savefig('process/colorselect.png')
````

## Comvert to Grayscale

````python
gray = []
for i in color_Select:
    temp = helper.grayscale(i)
    gray.append(temp)

plt.figure()
plt.title('Grayscale')
plt.imshow(gray[0],cmap='gray')
plt.savefig('process/grayscale.png')
````

## Define a kernel size of Gaussian blurring

````python
kernel_size = 5 # Must be an odd number(3, 5, 7...)
blur_gray = []
for i in gray:
    temp = helper.gaussian_blur(i, kernel_size)
    blur_gray.append(temp)

plt.figure()
plt.title('Gaussian blur')
plt.imshow(blur_gray[0],cmap='gray')
plt.savefig('process/gussian_blur.png')
````

## Canny Edges

````python
canny_edges = []
low_threshold = 250 
high_threshold = 300
for i in blur_gray:
    temp = helper.canny(i, low_threshold, high_threshold)
    canny_edges.append(temp)

plt.figure()
plt.title('Edges')
plt.imshow(canny_edges[0],cmap='gray')
plt.savefig('process/edges.png')
````

## Mask Area Selection

````python
masked_edges = []

for i in canny_edges:
    imshape = i.shape
    vertices = np.array([[(100,imshape[0]), #left_bottom
                          (470, 300), # left_top
                          (510, 300), # right_top
                          (imshape[1],imshape[0])]], # right_bottom
                        dtype=np.int32)
    temp = helper.region_of_interest(i, vertices)
    masked_edges.append(temp)

plt.figure()
plt.title('Masked Edges')
plt.imshow(masked_edges[0],cmap='gray')
plt.savefig('process/masked_edges.png')

plt.figure()
plt.title('Mask Selection')
plt.imshow(image[0])

left_bottom = [100,imshape[0]]
right_bottom = [imshape[1],imshape[0]]
left_top = [470, 300]
right_top = [510,300]

x = [left_bottom[0], right_bottom[0], right_top[0], left_top[0],left_bottom[0]]
y = [left_bottom[1], right_bottom[1], right_top[1], left_top[1],left_bottom[1]]
plt.plot(x, y, 'b--')
````

## Hough Transformation Line

````python
rho = 1
theta = np.pi/360
threshold = 4
min_line_len = 15 
max_line_gap = 4

hough_line = []
for i in masked_edges:
    temp = helper.hough_lines(i, rho, theta, threshold, min_line_len, max_line_gap)
    hough_line.append(temp)
    
plt.figure() 
plt.title("Hough Line Detection") 
plt.imshow(hough_line[0])
plt.savefig('process/hough_line.png')
````

## Result

````python
final = []
for i in range(len(hough_line)):
    temp = helper.weighted_img(hough_line[i], image[i], α=0.8, β=1., γ=0.)
    final.append(temp)
for i in range(len(image)):
    plt.subplot(2,3,i+1)
    plt.imshow(final[i])
````

````python
# Import everything needed to edit/save/watch video clips
from moviepy.editor import VideoFileClip
from IPython.display import HTML
````

````python
def process_image(image):
    # NOTE: The output you return should be a color image (3 channel) for processing video below
    # TODO: put your pipeline here,
    # you should return the final output (image where lines are drawn on lanes)
    
    #color select
    red_threshold = 200
    green_threshold = 160
    blue_threshold = 0
    colorselect = helper.colorselect(image, red_threshold, green_threshold, blue_threshold)
    
    # grayscale
    gray_img = helper.grayscale(colorselect)
    
    # Kernal size
    kernel_size = 5 # Must be an odd number(3, 5, 7...)
    blur_gray = helper.gaussian_blur(gray_img, kernel_size)
    
    # Canny edges
    low_threshold = 50
    high_threshold = 150
    canny_edges = helper.canny(blur_gray, low_threshold, high_threshold)
    
    # Mask Area
    imshape = canny_edges.shape
    vertices = np.array([[(100,imshape[0]), #left_bottom
                          (470, 300), # left_top
                          (510, 300), # right_top
                          (imshape[1],imshape[0])]], # right_bottom
                        dtype=np.int32)
    masked_edges = helper.region_of_interest(canny_edges, vertices)

    # Hough Line
    rho = 1
    theta = np.pi/360
    threshold = 4
    min_line_len = 15 
    max_line_gap = 5

    hough_line = helper.hough_lines(masked_edges, rho, theta, threshold, min_line_len, max_line_gap)
    
    result = helper.weighted_img(hough_line, image, α=0.8, β=1., γ=0.)
    
    return result
````
